```python
from cs336_basics.train import load_checkpoint, TrainConfig, build_model, train
from tests.test_tokenizer import get_tokenizer_from_vocab_merges_path, VOCAB_PATH, MERGES_PATH
import torch
```

```python
device="cuda:3"
src = '/sda1/szl/cs336/checkpoints/latest.pt'
cfg = TrainConfig(batch_size=32, d_model=640, context_length=256, train_tokens_path="", valid_tokens_path="", device=device, d_ff=1536, vocab_size=50257)
model = build_model(cfg)
model.load_checkpoint(src)
```

```python
tokenizer = get_tokenizer_from_vocab_merges_path(VOCAB_PATH, MERGES_PATH, special_tokens=["<|endoftext|>"])
text = "i like computer"
input_ids =torch.tensor([tokenizer.encode(text)])
output_ids = model.generate(input_ids, temperature=0.5).tolist()[0]
print(tokenizer.decode(output_ids))
```
